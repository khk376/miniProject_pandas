```python
%matplotlib inline
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
from glob import glob
```

```python
data = pd.read_csv("edu.csv", encoding = 'utf-8')
data
```

```python
del data['학교ID']
del data['학교명']
del data['설립일자']
del data['본교분교구분']
del data['운영상태']
del data['소재지도로명주소']
del data['시도교육청코드']
del data['시도교육청명']
del data['교육지원청코드']
del data['교육지원청명']
del data['생성일자']
del data['변경일자']
del data['위도']
del data['경도']
del data['데이터기준일자']
del data['설립형태']

data
```

```python
data
```

```python
data['동읍면']= [eachAddress.split()[2] for eachAddress in data['소재지지번주소']]
data
```

```python
data['동/읍/면'] = data['시군구명'].map(str) + " " + data['동읍면'].map(str)
data
```

```python
del data['동읍면']
```

```python
df=data.copy()
```

```python
df=df.groupby(['동/읍/면']).size().reset_index(name='학교 수')
```

```python
df
```

```python
df=df.rename(columns={"동/읍/면":"동읍면"})
df
```
